# Must Now Add Destination Stop ID & Time to Travel There

```python
df
```

```python
stop_df = df
```

```python
stop_df['TimeStamp'] =  pd.to_datetime(stop_df['TimeStamp'], format='%Y-%m-%d %H:%M:%S')
```

```python
stop_df.dtypes
```

```python
# List to hold addresses for destination addresses
dst_stop = list()
travel_time = list()

for item, row in stop_df.iterrows():
    
    # Set up values on first iteration
    if item == 0:
        last_stop = row[4]
        last_id = row[3]
        last_time = row[0]
        continue
    
    current_id = row[3]
    
    # If it's a new Journey ID
    if current_id != last_id:
        last_stop = row[4]
        last_id = row[3]
        last_time = row[0]
        df.drop(item - 1, inplace = True)
        continue
        
    dst_stop.append(row[4])
    travel_time.append(row[0] - last_time)
    last_time = row[0]
```

## Add Day Of Week And Modify TimeStamp

```python
stop_df['weekday'] = stop_df['TimeStamp'].dt.dayofweek
```

```python
stop_df['TimeStamp'] = stop_df['TimeStamp'].astype(str)
```

```python
for index, row in stop_df.iterrows():
    x = row["TimeStamp"][-8:]
    stop_df.set_value(index, 'TimeStamp', x)    
```

```python
# Convert to datetime
# ad_df['TimeStamp'] =  pd.to_datetime(ad_df['TimeStamp'], format='%H:%M:%S')
stop_df['TimeStamp'] = stop_df['TimeStamp'].astype(str)
# Convert to category
# ad_df["TimeStamp"] = ad_df["TimeStamp"].astype("category")
stop_df["weekday"] = stop_df["weekday"].astype("category")
```

```python
# Must convert travelTime into an int representing journey time in sec
import numpy as np

stop_df['TravelTime'] = stop_df['TravelTime'] / np.timedelta64(1, 's')
```

# Make Model For Address DF With StatsModels

```python
# Import package matplotlib for visualisation/plotting
import matplotlib.pyplot as plt
import matplotlib as mpl

# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

#For showing plots directly in the notebook
%matplotlib inline
```

```python
# create a fitted model with the features selected from exercise 1
lm = smf.ols(formula='travelTime ~ LineId + Address + dstAddress + TimeStamp + Rain + weekday', data = stop_df).fit()
```

```python
# Make small test
X_new = pd.DataFrame({'LineId': [145], 'Address':['Bray Road'],'dstAddress':['Shankill'], 'TimeStamp': ["10:00:05"], 'Rain':[0.0],'weekday':[1]})
X_new.head()
```
